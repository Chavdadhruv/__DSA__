```python
# Part 1: Array Creation
def create_array(elements):
    """Create an array from a list of elements."""
    return list(elements)

# Part 2: Insert an Element
def insert_element(arr, index, element):
    """Insert an element at a specific index."""
    arr.insert(index, element)
    return arr

# Part 3: Get an Element
def get_element(arr, index):
    """Retrieve an element from a specific index."""
    if 0 <= index < len(arr):
        return arr[index]
    else:
        return "Index out of range"

# Part 4: Search for an Element
def search_element(arr, element):
    """Search for an element in the array and return its index."""
    try:
        return arr.index(element)
    except ValueError:
        return "Element not found"

# Part 5: Delete an Element
def delete_element(arr, element):
    """Delete an element from the array."""
    try:
        arr.remove(element)
        return arr
    except ValueError:
        return "Element not found"

# Part 6: Filter Array
def filter_array(arr, condition):
    """Filter elements in the array based on a condition."""
    return [element for element in arr if condition(element)]

# Part 7: Merge Arrays
def merge_arrays(arr1, arr2):
    """Merge two arrays."""
    return arr1 + arr2

# Part 8: Reverse Array
def reverse_array(arr):
    """Reverse the array."""
    return arr[::-1]

# Part 9: Rotate Array
def rotate_array(arr, k):
    """Rotate the array by k steps."""
    k %= len(arr)  # Handle rotation greater than array length
    return arr[-k:] + arr[:-k]

# Part 10: Additional Array Functions

# Sum of all elements in the array
def sum_array(arr):
    """Return the sum of all elements in the array."""
    return sum(arr)

# Find the maximum element in the array
def max_element(arr):
    """Return the maximum element in the array."""
    return max(arr)

# Find the minimum element in the array
def min_element(arr):
    """Return the minimum element in the array."""
    return min(arr)


# Example Usage:
if __name__ == "__main__":
    # Creating an array
    arr = create_array([10, 20, 30, 40, 50])
    print("Array:", arr)
    
    # Inserting an element
    arr = insert_element(arr, 2, 25)
    print("After Insertion:", arr)
    
    # Getting an element
    element = get_element(arr, 3)
    print("Element at index 3:", element)
    
    # Searching for an element
    index = search_element(arr, 40)
    print("Index of element 40:", index)
    
    # Deleting an element
    arr = delete_element(arr, 20)
    print("After Deletion:", arr)
    
    # Filtering the array (keeping even numbers)
    filtered_arr = filter_array(arr, lambda x: x % 2 == 0)
    print("Filtered Array (even numbers):", filtered_arr)
    
    # Merging two arrays
    arr2 = create_array([60, 70, 80])
    merged_arr = merge_arrays(arr, arr2)
    print("Merged Array:", merged_arr)
    
    # Reversing the array
    reversed_arr = reverse_array(arr)
    print("Reversed Array:", reversed_arr)
    
    # Rotating the array by 2 steps
    rotated_arr = rotate_array(arr, 2)
    print("Rotated Array (by 2 steps):", rotated_arr)
    
    # Sum of array elements
    array_sum = sum_array(arr)
    print("Sum of elements in the array:", array_sum)
    
    # Maximum element in the array
    max_val = max_element(arr)
    print("Maximum element in the array:", max_val)
    
    # Minimum element in the array
    min_val = min_element(arr)
    print("Minimum element in the array:", min_val)
```
